```python
import bs4, requests

# 수집할 URL정보만 수정
# targetUrl = "https://sparkkorea.com/%ed%80%b4%ec%a6%88/"
targetUrl = "https://finance.naver.com/"

# 웹접속 시 TRY EXCEPT 적용
try :
    resp = requests.get(targetUrl)
except Exception as e:
    print(e)

webPageSrc = resp.text
htmlObj = bs4.BeautifulSoup(webPageSrc, "html.parser")
```
