```python
#원하는 사이트에서 정보 받아오기

import requests
import bs4
import pandas as pd

targetUrl = "https://www.ssg.com/search.ssg?target=all&query=%EC%BB%A4%ED%94%BC"

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.106 Safari/537.36',}

indata = requests.get(targetUrl, headers=headers)

ssg=indata.text

htmlObj = bs4.BeautifulSoup(ssg, "html.parser")

htmlObj
```

```python
# ssg.com 커피 1페이지 첫번째 커피 이름과 가격정보 수집 (광고제외)

areaObj = htmlObj.find(id="area_searchItemList")

li = areaObj.find(name = "li")

unitInfo = li.find(attrs = {"class" : "cunit_info"})

coffee_price = unitInfo.findAll(name = "em", attrs = {"class" : "ssg_price"})

coffee_price

coffee_name = unitInfo.findAll(name = "em", attrs = {"class" : "tx_ko"})

coffee_name

len(coffee_name)

coffeInfoOutput = []

for i in range(0, len(coffee_name)) :
    coffeInfoOutput.append(coffee_name[i].text)
    coffeInfoOutput.append(coffee_price[i].text)
     
coffeInfoOutput        
```
