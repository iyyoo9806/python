```python
def productAgeDefine(inValue):
    # inValue = 1.5
    
    outValue = 1
    if inValue < 1:
        outValue = 1
    elif inValue < 2:
        outValue = 2
    elif inValue < 3:
        outValue =3
    else:
        outValue = 4

    return outValue
```

```python
customerData["PRODUCTAGE_NEWF"] = customerData.PRODUCTAGE.apply(productAgeDefine)
```
