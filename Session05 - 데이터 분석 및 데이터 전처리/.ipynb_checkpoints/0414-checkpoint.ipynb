```python
import pandas as pd
import numpy as np

customerData = pd.read_csv("../dataset/customerdata.csv")
```

```python
# <1 이면 1: 그 외에는 기존값.
customerData["PRODUCTAGE_NEW"] = np.where(customerData.PRODUCTAGE <1,1,0)
```

```python
customerData["PRODUCT_NEW_AGE"] = \
np.where(customerData.PRODUCTAGE<1,1,
         np.where(customerData.PRODUCTAGE <2,2,
                  np.where(customerData.PRODUCTAGE<3,3,4)))
```

```python
inValue = 1.5
outValue = 1
```

```python
def productAgeDefine(inValue):
    # inValue = 1.5
    
    outValue = 1
    if inValue < 1:
        outValue = 1
    elif inValue < 2:
        outValue = 2
    elif inValue < 3:
        outValue =3
    else:
        outValue = 4

    return outValue
```

```python
customerData["PRODUCTAGE_NEWF"] = customerData.PRODUCTAGE.apply(productAgeDefine)
```
