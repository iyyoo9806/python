## 주차별 효과 생성

```python
selloutData = pd.read_csv("../dataset/kopo_channel_seasonality_new.csv")
```

```python
sortKey = ["REGIONID","PRODUCT","YEARWEEK"]
sortedData = selloutData.sort_values(by=sortKey,ascending=[True,True,True])
```

```python
sortedData["MA5"] = sortedData["QTY"].rolling(window=5,center=True, min_periods=1).mean()
```

```python
sortedData["SEASONALITY"] = sortedData["QTY"] / sortedData["MA5"]
```

## 연도 컬럼 생성

```python
sortedData["YEAR"] = sortedData.YEARWEEK.astype(str).str[0:4]
sortedData["WEEK"] = sortedData.YEARWEEK.astype(str).str[4:]
```

## 그룹바이 그룹키(지역,상품,주차)

```python
groupKey = ["REGIONID","PRODUCT","WEEK"]
```

```python
firstData = sortedData.groupby(by=groupKey)["SEASONALITY"].agg(["mean"]).reset_index()
firstData.rename(columns={"mean":"SEASONALITY"},inplace=True)
```

## INNER JOIN

```python
### 교집합 영역만
```

```python
row1 = ["A01","ST0001",0.3]
row2 = ["A02","ST0001",0.5]
row3 = ["A02","ST0002",0.5]
row4 = ["A99","ST0099",0.9] 
promoData = pd.DataFrame([ row1,row2,row3], columns = ["REGION","PD","PROMOTION"])
promoData
```

```python
pd.merge(left = joinSellout,
        right = promoData,
        left_on=["SALESID","PRODUCT"],
        right_on=["REGION","PD"])
```

## LEFT JOIN

```python
### 기준이 되는 데이터는 무조건 살림!
```

```python
pd.merge(left = joinSellout,
        right = promoData,
        left_on=["SALESID","PRODUCT"],
        right_on=["REGION","PD"],
        how = "left")
```

# RIGHT JOIN

```python
pd.merge(left = joinSellout,
        right = promoData,
        left_on=["SALESID","PRODUCT"],
        right_on=["REGION","PD"],
        how = "right")
```

# OUTER JOIN

```python
##LEFT RIGHT 모두
pd.merge(left = joinSellout,
        right = promoData,
        left_on=["SALESID","PRODUCT"],
        right_on=["REGION","PD"],
        how = "outer")
```

# CROSS(안씀)

```python
pd.merge(left = joinSellout,
         right = promoData,
         how = "cross")
```
