## A-1 타입통합 및 불량 데이터 처리

```python
import pandas as pd
import numpy as np
```

```python
selloutData = pd.read_csv("../dataset/kopo_channel_seasonality_new.csv")
```

```python
selloutData
```

```python
selloutData.info()
```

```python
# 타입 통합
```

```python
selloutData.QTY = selloutData.QTY.astype(float)
selloutData.REGIONID = selloutData.REGIONID.astype(str)
selloutData.PRODUCT = selloutData.PRODUCT.astype(str)
selloutData.YEARWEEK = selloutData.YEARWEEK.astype(str)
```

```python
selloutData.info()
```

```python
selloutData
```

```python
# 불량 데이터 처리(음수 데이터를 양수로 처리한다.)
```

```python
selloutData["QTY_NEW"] = \
np.where(selloutData.QTY<0,0,selloutData.QTY)
```

```python
selloutData.loc[(selloutData.QTY_NEW<0)&(selloutData.QTY_NEW!=0)]
```

```python
selloutData
```

## A-2 데이터 통합

```python
selloutData["YEAR"] = selloutData.YEARWEEK.str[:4]
```

```python
selloutData["WEEK"] = selloutData.YEARWEEK.str[4:]
```

```python
selloutData.WEEK = selloutData.WEEK.astype(int)
```

```python
refinedSelloutData = selloutData.loc[selloutData.WEEK<53]
```

```python
refinedSelloutData
```

```python
refinedSelloutData.loc[refinedSelloutData.WEEK>52]
```

```python
refinedSelloutData 

# PANDAS 패키지 불러오기
import pandas as pd 
import pymysql 
from sqlalchemy import create_engine 

inId = "kopo"
inPw = "kopo"
inIp = "192.168.110.111"
inPort ="3306"
inDbName = "kopo"

# DB 커넥션 열기
engine_mariadb = create_engine('mysql+pymysql://{}:{}@{}:{}/{}'.format(inId,inPw,inIp,inPort,inDbName))

refinedSelloutData.to_sql(name="refinedSelloutData_유인영", con= engine_mariadb, if_exists="replace", index = False)
```

```python
refinedSelloutData
```

```python
sortKey = ["REGIONID","PRODUCT","YEARWEEK"]
```

```python
sortedData = refinedSelloutData.sort_values(by=sortKey, ascending=[True,True,True])
```

```python
sortedData
```

```python
groupKey =  ["REGIONID","PRODUCT","YEAR"]
firstData = sortedData.groupby(by = groupKey)["QTY_NEW"].agg(["mean"])
```

```python
firstData.rename(columns={"mean":"QTY_MEAN"},inplace=True)
```

```python
groupData=firstData.reset_index()
```
